1. <span style="background-color: rgb(255, 255, 255); font-size: 10pt; white-space: pre-wrap; color: rgb(102, 102, 102); font-family: Arvo, serif;">How many employees were reported in 2016 in all industries? ```sql
use LaborStatisticsDB
go

select round(sum(value),0) as "Number of employees reported in 2016"
from annual_2016
inner join series on series.series_id = annual_2016.series_id
inner join datatype on datatype.data_type_code = series.data_type_code
where datatype.data_type_text = 'all employees'
```

4. <span style="background-color: rgb(255, 255, 255); font-size: 10pt; white-space: pre-wrap; color: rgb(102, 102, 102); font-family: Arvo, serif;">In January 2017, what is the average weekly hours worked by production and nonsupervisory employees across all industries?</span>

5. <span style="background-color: rgb(255, 255, 255); font-size: 10pt; white-space: pre-wrap; color: rgb(102, 102, 102); font-family: Arvo, serif;">What is the total weekly payroll for production and nonsupervisory employees across all industries in January 2017? ```sql
use LaborStatisticsDB
go

select round(sum(value),2) as "Total weekly payroll of production/nonsupervisory employees in January 2017"
from january_2017
inner join series on series.series_id = january_2017.series_id
inner join datatype on datatype.data_type_code = series.data_type_code
where datatype.data_type_text = 'average weekly earnings of production and nonsupervisory employees'
```

6. <span style="background-color: rgb(255, 255, 255); font-size: 10pt; white-space: pre-wrap; color: rgb(102, 102, 102); font-family: Arvo, serif;">In January 2017, which industry was the average weekly hours worked by production and nonsupervisory employees the highest? Which industry was the lowest?</span>

```sql
use LaborStatisticsDB
go

select top 1 industry_name as "Industry with Highest Average Weekly Hours for Production and Nonsupervisory Employees"
from industry
inner join series on series.industry_code = industry.industry_code
inner join january_2017 as j on j.series_id = series.series_id
inner join datatype on datatype.data_type_code = series.data_type_code
where data_type_text = 'average weekly hours of production and nonsupervisory employees' and industry.industry_name is not null
group by industry.industry_name
order by avg(j.value) desc;

select top 1 industry_name as "Industry with Lowest Average Weekly Hours for Production and Nonsupervisory Employees"
from industry
inner join series on series.industry_code = industry.industry_code
inner join january_2017 as j on j.series_id = series.series_id
inner join datatype on datatype.data_type_code = series.data_type_code
where data_type_text = 'average weekly hours of production and nonsupervisory employees' and industry.industry_name is not null
group by industry.industry_name
order by avg(j.value) asc;
```

1. <span style="background-color: rgb(255, 255, 255); font-size: 10pt; white-space: pre-wrap; color: rgb(102, 102, 102); font-family: Arvo, serif;">Write a query that returns the series_id, industry_code, industry_name, and value from the january_2017 table but only if that value is greater than the average value for annual_2016 of data_type_code 82.</span>

```sql
use LaborStatisticsDB
go

select distinct j.series_id, j.value, i.industry_code, i.industry_name
from january_2017 as j
inner join series on series.series_id = j.series_id
inner join industry as i on i.industry_code = series.industry_code
where j.value > (select avg(a.value)
    from annual_2016 as a
    inner join series on series.series_id = a.series_id
    inner join datatype on datatype.data_type_code = series.data_type_code
    where datatype.data_type_code = 82)
```

2. <span style="background-color: rgb(255, 255, 255); font-size: 10pt; white-space: pre-wrap; color: rgb(102, 102, 102); font-family: Arvo, serif;">Create a Union table comparing average weekly earnings&nbsp; of production and nonsupervisory employees between annual_16 and january_17 using the data type 30.&nbsp; Round to the nearest penny.&nbsp; You should have a column for the average earnings and a column for the year, and the period.</span>

```sql
use LaborStatisticsDB
go

select a.period, a.year, round(avg(a.value),2) as "Average weekly earnings of production/nonsupervisory employees"
from annual_2016 as a
inner join series on series.series_id = a.series_id
inner join datatype on datatype.data_type_code = series.data_type_code
where datatype.data_type_code = 30
group by a.period, a.year
union all
select j.period, j.year, round(avg(j.value),2) as "Average weekly earnings of production/nonsupervisory employees"
from january_2017 as j
inner join series on series.series_id = j.series_id
inner join datatype on datatype.data_type_code = series.data_type_code
where datatype.data_type_code = 30
group by j.year, j.period;
```
